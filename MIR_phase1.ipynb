```python
punctuations = ['؟', '.','،', '؛', '!', ':',')','(', '\n', '<', '>', '-', '[',']', '«', '»', '"']
# english = False
english = True
```

```python
def get_stop_words():
    """Detects stop-words

    Returns
    -------
    dict
        A dictionary of terms with their repetition
    """
    if english:
        data = []
        for i in range(0, len(text_english)):
            tokens = nltk.word_tokenize(text_english.values[i][1])
            tokens = [word.lower() for word in tokens if word.isalnum()]
            data += tokens
            tokens = nltk.word_tokenize(text_english.values[i][2])
            tokens = [word.lower() for word in tokens if word.isalnum()]
            data += tokens
    else:
        data = []
        for i in range(0, len(text_persian)):
            tokens = hazm.word_tokenize(text_persian.values[i][0])
            tokens = [word for word in tokens if  not (word in punctuations)]
            data += tokens
            tokens = hazm.word_tokenize(text_persian.values[i][1])
            tokens = [word for word in tokens if  not (word in punctuations)]
            data += tokens

    data_analysis = nltk.FreqDist(data)
    output = dict()
    remove_frequency = (len(data)) // 200
    for key in data_analysis:
        if data_analysis[key] >= remove_frequency:
            output[key] = data_analysis[key] 
      
    return output
  

stop_words = get_stop_words()
stop_words
```

```python
def prepare_text(raw_text):
  """Preprocesses the text with tokenization, case folding, stemming and lemmatization

  Parameters
  ----------
  raw_text : str
      The title or plot of a movie

  Returns
  -------
  list
      A list of tokens
  """
  if english:
    # ------------ tokenization
    tokens = nltk.word_tokenize(raw_text)
    
    # remove punctuation
    tokens = [word for word in tokens if word.isalnum()]

    # ------------ stopword removal
    tokens = [word for word in tokens if not (word in stop_words.keys())]

    # ------------ lemmatizing
    wn_lemm= nltk.WordNetLemmatizer()
    tokens = [wn_lemm.lemmatize(w, 'v') for w in tokens]

    # ------------ stemming
    ps = PorterStemmer()
    ss = SnowballStemmer('english')
    ls = LancasterStemmer()
    # rs = RegexpStemmer('ing$|s$|ed$|y$', min=2) # set the minimum of the string to stem
    tokens = [ps.stem(w) for w in tokens]
    tokens = [ss.stem(w) for w in tokens]
    tokens = [ls.stem(w) for w in tokens]
    
    return tokens#["edvard", "am", "a", "run", "he", "am", "always", "run"]
  else:
    # remove half space
    raw_text = raw_text.replace("‌", " ")

    # ------------ normalizing
    normalizer = hazm.Normalizer()
    tokens = normalizer.normalize(raw_text)

    # ------------ tokenization
    tokens = hazm.word_tokenize(tokens)

    # remove punctuations
    tokens = [word for word in tokens if  not (word in punctuations)]

    # ------------ stopword removal
    tokens = [word for word in tokens if not (word in stop_words.keys())]

    # ------------ stemming
    stemmer = hazm.Stemmer()
    tokens = [stemmer.stem(w) for w in tokens]

    # ------------ lemmatizing
    lemmatizer = hazm.Lemmatizer()
    tokens = [lemmatizer.lemmatize(w) for w in tokens]
    tokens = [w for w in tokens if w != '']
    

    return tokens

prepare_text("Edvard ate a CAKE? He was always running.")
# prepare_text('اصلاح نویسه ها و استفاده از نیم فاصله ، پردازش را آسان می‌کند')
```

```python
from bitarray import bitarray

def variable_byte_code(number):
    encoded = ''
    first_flag = True
    while True:
        lsb = number & 127
        number >>= 7
        if first_flag:
            encoded = '1'+ str(format(lsb, 'b')).zfill(7)
            first_flag = False
            if number == 0:
                break
            continue
        encoded = '0'+ str(format(lsb, 'b')).zfill(7) + encoded
        if number == 0:
            break
    return int(encoded)

def gamma_code(number):
    if number == 0:
        return None
    binary_n = format(number, 'b')
    binary_offset = binary_n[1::]
    unary_offset_length = bitarray(True for i in range(len(binary_offset))) + bitarray([False])
    gamma_str = ''
    for b in (unary_offset_length+binary_offset):
        if b==True:
            gamma_str += '1'
        else:
            gamma_str += '0'
    return int(gamma_str)
```

<h3>
توضیح(مربوط به بخش سوم)  
</h3>
برای فشرده‌سازی نمایه‌ها دو روش gamma_code و variable_byte_code استفاده می‌کنیم. تابع‌های  decode هر دو در ادامه توضیح داده می‌شود.
</br>
 variable_byte_code:  این تابع عدد را گرفته و با واحد بایت یعنی تعداد بیت مضرب 8 آن را کد میکند.
    میتوان گفت فرم باینری آن را نوشته و پس از 7 بیت کم ارزش یک بیت 1 و پس از 7 بیت سایر بایت‌ها، عدد 0 را  قرار میدهیم.
    از and بیتی با 127 و اپراتور شیفت به راست استفاده شده است.
</br>
 gamma_code:   فرم باینری عدد را نوشته و msb آن را حذف میکند. سپس آن را آفست نامیده و به اندازه‌ی طول آفست یک گذاشته سپس 0 قرار میدهد و سپس آفست را نیز در سمت راست عدد قرار میدهد.      
</div>

```python
import json
import os

def store_index(tokens_of_docs, path, compression_type):
    """Stores the index in a file

    Parameters
    ----------
    path : str
        Path to store the file

    compression_type : str
        Could be one of the followings:
        - no-compression
        - gamma-code
        - variable-byte

    Returns
    -------
    int
        The size of the stored file
    """
    
    #create positional_index (no-compression)
    positional_index = {}
    for docID in range(len(tokens_of_docs)):
        for position, term in enumerate(tokens_of_docs[docID][0]):
            if term in positional_index:
                if docID in positional_index[term][1]:
                    positional_index[term][1][docID].append(position)      
                else:
                    positional_index[term][0] = positional_index[term][0] + 1
                    positional_index[term][1][docID] = [position]
            else:
                positional_index[term] = []
                positional_index[term].append(1)
                positional_index[term].append({})   #create postings list  
                positional_index[term][1][docID] = [position]
    
    #sorting
    positional_index = dict(sorted(positional_index.items()))
    for term in positional_index:
        positional_index[term][1] = dict(sorted(positional_index[term][1].items()))
   

    if compression_type == 'no-compression':
        with open(path, 'w') as fp:
            json.dump(positional_index, fp)
          
        size_of_file = os.path.getsize(path)
        return size_of_file


    #compression    
    new_positional_index = {}
    for term in positional_index:
        new_positional_index[term] = []
        new_positional_index[term].append(positional_index[term][0])
        new_positional_index[term].append([])
        preID = 0
        for docID in positional_index[term][1]:
            new_docID = docID - preID
            if compression_type == 'variable-byte':
                new_docID = variable_byte_code(new_docID)
            elif compression_type == 'gamma-code':
                new_docID =  gamma_code(new_docID)
            new_positional_index[term][1].append([new_docID, positional_index[term][1][docID].copy()])
            preID = docID                
    
    
 #store positional_index 
    with open(path, 'w') as fp:
        json.dump(new_positional_index, fp)
    
    size_of_file = os.path.getsize(path)
    return size_of_file


tokens_of_docs = []
if english:
    prefix = 'en_'
    for i in range(0, len(text_english)):
        tokens_of_docs.append([])
        tokens_of_docs[-1].append(prepare_text(text_english.values[i][1]))
        tokens_of_docs[-1].append(prepare_text(text_english.values[i][2]))  
else:
    prefix = 'fa_'
    for i in range(0, len(text_persian)):
        tokens_of_docs.append([])
        tokens_of_docs[-1].append(prepare_text(text_persian.values[i][0]))
        tokens_of_docs[-1].append(prepare_text(text_persian.values[i][1]))

print(str(store_index(tokens_of_docs, prefix+"json_no_compression.json", "no-compression")) + "\tbytes")
print(str(store_index(tokens_of_docs, prefix+"json_variable_byte.json", "variable-byte")) + "\tbytes")
print(str(store_index(tokens_of_docs, prefix+"json_gamma_code.json", "gamma-code")) + "\tbytes")
```

<h3>
توضیح (مربوط به بخش دوم)  
</h3>
برای پیاده‌سازی نمایه جایگاهی(positional index)، دیکشنری‌ای میسازیم و در آن به ازای هر term، تعداد داک‌های شامل آن لغت و لیستی از آیدی داک ها به همراه جایگاه هایی که این لغت در داک آمده است را نگهداری میکنیم.
در ادامه به توضیح هر تابع میپردازیم:
</br>
store_index: از این تابع برای ساخت نمایه جایگاهی، فشرده سازی آن و ذخیره آن در فایل استفاده میکنیم.
لیست شامل توکن‌های هر داک به عنوان آرگومان ورودی داده میشود. سپس به ازای هر سند، باید جفت های (لغت , جایگاه) آن را گرفته و اگر آن لغت در نمایه وجود نداشته باشد(لغت جدید)، برای آن یک جایگاه در دیکشنری میسازیم و اگر موجود باشد، جایگاه وقوع لغت و در صورت جدید بودن شماره داک را هم به posting list این لغت اضافه میکنیم.
پس از این مرحله، term ها و همچنین posting list هر ترم را مرتب‌سازی میکنیم.
اگر فشرده‌سازی نیاز است، به روش خواسته شده نمایه را فشرده میکنیم. و در آخر  positional index ساخته شده را در در فایل .json ذخیره میکنیم. 
    برای فشرده سازی نیز از تابع های variable_byte_code  و gamma_code استفاده شده که قبل‌تر، آن‌ها را توضیح دادیم.
    فشرده‌سازی در این مرحله باعث کاهش قابل توجه حجم posting list ها میشود.
</div>

```python
def variable_byte_decode(code):
    n = ''
    fromm = 1
    to = 8
    code = str(code)
    while code[fromm-1] == '0':
        n += code[fromm:to]
        fromm += 8
        to += 8
    n += code[-7:]
    return int(n,2)


def gamma_decode(code):
    if code == '' or code == None:
        return 0
    code = str(code)
    lenn  = str(code).find('0')
    offset = str(code)[-lenn:]
    return int(offset,2) + 2 ** lenn
```

<h3>
توضیح(مربوط به بخش سوم)  
</h3>
 variable_byte_decode: برای دیکود کردن ابتدا بیت‌های هشتم را حذف کرده و عدد باقی مانده را به دسیمال تبدیل میکند. انتهای عدد را نیز از 1 بودن بیت هشتم آخرین بایت تشخیص می‌دهد.
</br>
 gamma_decode: جهت دیکود کردن کد گاما استفاده شده و با یافتن جایگاه اولین صفر، طول آفست را فهمیده و از سمت راست offset را میخواند. 1 پرارزش را اضافه کرده و سپس کل عدد باینری را به دسیمال تبدیل می‌کند.
</div>

```python
def load_index(path, compression_type):
    """Loads the index from a file

    Parameters
    ----------
    path : str
        Path of the file to load from

    compression_type : str
        Could be one of the followings:
        - no-compression
        - gamma-code
        - variable-byte
    """

    #load
    with open(path) as json_file:
        data = json.load(json_file)
    
    #decompress positional_index
    if compression_type == 'no-compression':
        return data
    
    positional_index = {}
    for term in data:
        positional_index[term] = []
        positional_index[term].append(data[term][0])
        positional_index[term].append({})
        preID = '' # for 'gamma-code'
        if compression_type == 'variable-byte':
            preID = '10000000' # for variable-byte-code
        for pair in data[term][1]:
            new_docID =  gamma_decode(preID) + gamma_decode(pair[0]) # for 'gamma-code'   
            if compression_type == 'variable-byte':
                new_docID = variable_byte_decode(preID) +  variable_byte_decode(pair[0])
            
            positional_index[term][1][new_docID] = pair[1]
            preID = pair[0] 
        
    return positional_index

if english:
    prefix = 'en_'
else:
    prefix = 'fa_'

positional_index = load_index(prefix+"json_no_compression.json", "no-compression")
# print(positional_index)
positional_index = load_index(prefix+"json_variable_byte.json", "variable-byte")
print(positional_index)
positional_index = load_index(prefix+"json_gamma_code.json", "gamma-code")
# print(positional_index)
```

توضیح (مربوط به بخش دوم)  
</h3>
load_index: این تابع ابتدا، positional index را از فایل .json لود کرده و اگر نمایه، فشرده باشد، آن را decode میکند. برای اینکار از تابع های gamma_decode و variable_byte_decode استفاده میکنیم که توضیح آن‌ها نیز در قبل آمده است.
</div>

```python
def delete_doc(docID, positional_index): #no-compression
    for term in positional_index:
        if docID in positional_index[term][1]:
            positional_index[term][0] -= 1
            del positional_index[term][1][docID]
    for k, v in list(positional_index.items()):
        if v[0] == 0:
            positional_index.pop(k)           
    return positional_index


def add_doc(tokens_of_new_doc, new_docID, positional_index): #no-compression
    for position, term in enumerate(tokens_of_new_doc):
            if term in positional_index:
                if new_docID in positional_index[term][1]:
                    positional_index[term][1][new_docID].append(position)      
                else:
                    positional_index[term][0] = positional_index[term][0] + 1
                    positional_index[term][1][new_docID] = [position]
            else:
                positional_index[term] = []
                positional_index[term].append(1)
                positional_index[term].append({})   #create postings list  
                positional_index[term][1][new_docID] = [position]
    
    #sorting
    positional_index = dict(sorted(positional_index.items()))
    for term in positional_index:
        positional_index[term][1] = dict(sorted(positional_index[term][1].items()))
        
    return positional_index  


if english:
    item = len(text_english)
    tokens = ['salam','new','ab','ab','cd','hi']
else:
    item = len(text_persian)
    tokens = ['سلام', 'جدید', 'بای']
    
print('Initial document')
print(positional_index)
# item = int(input('insert your item number'))
print('Adding document...')
add_doc(tokens, item, positional_index)
print(positional_index)
print('Deleting document...')
delete_doc(item, positional_index)
print(positional_index)
```

نیازی به ذخیره‌سازی و فشرده‌سازی نمایه بایگرم نیست. همچنین می‌توانید از کد آماده برای محاسبه edit distance استفاده کنید.

```python
from nltk.util import ngrams

def create_bigram_index():
    """Creates the bigram index for spell correction
    
    Returns
    -------
    bigram : set
        Bigram index of the dataframe.
    """
    bigram = {}

    # bigramming
    if english:
        df, title, text = text_english, 'title', 'text'
    else:
        df, title, text = text_persian, 'عنوان مقاله', 'توضیحات مقاله'
    
    for index, row in df.iterrows():
        title_, text_ = row[title], row[text]
        tokens = prepare_text(title_) + prepare_text(text_)
        for token in tokens:
            bichars = ngrams(token, 2)
            for bichar in bichars:
                bichar = bichar[0] + bichar[1]
                if bichar not in bigram:
                    bigram[bichar] = set()
                bigram[bichar].add(token)
    
    # set to list
    for k in bigram:
        bigram[k] = list(bigram[k])

    return bigram
    # bigram_index = {"he": ["hello", "hell", "he", "her"], "el": ["telephone", "else"]}

csv_bigram = create_bigram_index()
csv_bigram
```

توضیحات:
</h3>
تابع پیاده سازی شده وظیفه دارد تا تمامی بایگرم های مربوط به دیتافرم ورودی را تولید و در قالب یک دیکشنری ذخیره کند. با استفاده از تابع آماده ngrams که در اینجا باید n همان 2 باشد، تمامی دو کاراکتر های متوالی را در یک دیکشنری ذخیره می کند که مقدار این کلید، لیستی از کلماتی می باشد که این 2 کاراکتر پیوسته در آنها ظاهر شده است.
در انتها نیز مقدار هر کلید را که یک مجموعه است(برای جلوگیری از وجود کلمات تکراری) را نیز به لیست تبدیل می کنیم و آن را خروجی می دهیم.
</div>

```python
from nltk.metrics.distance import jaccard_distance, edit_distance
import math

def find_best_match(word, threshold=0.4):
    # bigramming a word in query
    candidates = set()
    bichars = ngrams(word, 2)
    for bichar in bichars:
        bichar = bichar[0] + bichar[1]
        if bichar in csv_bigram:
            for token in csv_bigram[bichar]:
                candidates.add(token)

    # choosing the best candidate based on jaccard and edit distance with threshold 0.4
    best_candidate, min_distance = word, math.inf
    for candidate in candidates:
        jaccard_dist = jaccard_distance(set(ngrams(candidate, 2)), set(ngrams(word, 2)))
        if jaccard_dist < threshold:
            edit_dist = edit_distance(candidate, word, transpositions=False)
            if edit_dist < min_distance:
                best_candidate, min_distance = candidate, edit_dist
        

    return best_candidate
```
